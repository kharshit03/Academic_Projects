#### Checking for outliers and removing them

```python
shop_df.columns
```

```python
# num_cols means numerical columns
num_cols = ['Administrative', 'Administrative_Duration', 'Informational', 'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration',
       'BounceRates', 'ExitRates', 'PageValues', 'SpecialDay']
# cat_cols means categorical columns
cat_cols = ['Month', 'OperatingSystems', 'Browser', 'Region', 'TrafficType', 'VisitorType', 'Weekend', 'Revenue']
```

```python
# Utility Function For outlier detection
def outlier_detect(df, colname, nrows = 2, mcols = 2,  width = 20, height = 15):
    fig, ax = plt.subplots(nrows, mcols, figsize=(width, height))
    fig.set_facecolor('lightgrey')
    rows = 0
    for var in colname:
        ax[rows][0].set_title('Boxplot for outlier detection ', fontweight = 'bold')
        plt.ylabel(var, fontsize=8)
        sns.boxplot(y = shop_df[var], color = 'g', ax = ax[rows][0])

        sns.histplot(shop_df[var], color='g', kde=True, ax = ax[rows][1])
        ax[rows][1].axvline(shop_df[var].mean(), color = 'r', linestyle = '--', label = 'Mean')
        ax[rows][1].axvline(shop_df[var].median(), color = 'g', linestyle= '-', label = 'Median')
        ax[rows][1].axvline(shop_df[var].mode()[0], color = 'royalblue', linestyle = '-', label = 'Mode')
        ax[rows][1].set_title("Outlier detection", fontweight = 'bold')
        ax[rows][1].legend({'Mean':shop_df[var].mean(), 'Median': shop_df[var].median(), 'Mode': shop_df[var].mode()})
        rows = rows + 1
    plt.show()
```

```python
def cat_bi_analysis(df, colname, depend_var, nrows = 2, mcols = 2, width = 20, height = 15):
    fig, ax = plt.subplots(nrows, mcols, figsize=(width, height))
    sns.set(style = 'white')
    rows = 0
    string = " based Distribution"
    for var in colname:
        string = var + string
        sns.countplot(data=shop_df, x = depend_var, hue = var, palette="hls", ax = ax[rows][0])
        sns.countplot(data=shop_df, x = var, hue = depend_var, palette="husl", ax = ax[rows][1])

        ax[rows][0].set_title(string, fontweight = "bold", fontsize = 10)
        ax[rows][1].set_title(string, fontweight = "bold", fontsize = 10)
        
        ax[rows][0].set_ylabel('count', fontweight = "bold", fontsize = 10)
        ax[rows][0].set_xlabel(var, fontweight = "bold", fontsize = 10)

        ax[rows][1].set_ylabel('count', fontweight = "bold", fontsize = 10)
        ax[rows][1].set_xlabel(var, fontweight = "bold", fontsize = 10)

        rows = rows + 1
        string = " based Distribution"
    plt.show()
```

```python
# Utility Function For Categorical Variable
def cat_analysis(df, colname, nrows = 2, mcols = 2, width = 20, height = 30, sortbyindex = False):
    fig, ax = plt.subplot(nrows, mcols, figsize=(width, height))
    fig.set_facecolor(color = 'white')
    string = "frequency of "
    rows = 0
    for colname in colnames:
        count = (shop_df[colname].value_counts(normalize = True) * 100)
        string = string + colname + ' in (%)'
        if sortbyindex:
            count = count.sort_index()
        count.plot.bar(color = sns.color_pallete("crest"), ax = ax[rows][0])
        ax[rows][0].set_ylabel(string, fontsize = 8)
        ax[rows][0].set_xlabel(colname, fontsize = 8)
        count.plot.pie(colors = sns.color_palette("crest"), autopct = '%0.0f%%',
        textprops = {'fontsize': 14}, ax = ax[rows][1])
        string = "Frequency of "
        rows = rows + 1 
```

```python
# num_cols means numerical columns
num_cols = ['Administrative', 'Administrative_Duration', 'Informational', 'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration',
       'BounceRates', 'ExitRates', 'PageValues', 'SpecialDay']
outlier_detect(shop_df, num_cols, 10, 2, 10, 50)
```
