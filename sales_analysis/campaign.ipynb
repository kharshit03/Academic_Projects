```python
def category(data):
    if 1 >= data <= 5:
        return 'low traffic'
    elif 6 >= data <= 10:
        return 'mid traffic'
    elif 11 >= data <= 15:
        return 'mid high traffic'
    else :
        return 'high traffic'

shopping['TrafficTypeCategory'] = shopping['TrafficType'].apply(category)
```
